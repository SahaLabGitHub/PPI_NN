```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the feature importance data
file_path = "/Users/qingshuzhao/Library/CloudStorage/OneDrive-SharedLibraries-UWM/Arjun Saha - qingshu_project/Paper/SI/model_evaluation/504_training_result/feature_importance.csv"
df = pd.read_csv(file_path)

# Check column names
print(df.columns)

feature_col = "Feature"  
shap_col = "Importance (SHAP value)" 

# Sort features by absolute SHAP value and select top 10
df = df.sort_values(by=shap_col, key=abs, ascending=False)
df_top10 = df.head(10)

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.barh(df_top10[feature_col], df_top10[shap_col], color="skyblue")
plt.xlabel(" Impact on Model Output (SHAP Value)",fontsize=17)
plt.yticks(fontsize=14) 
plt.xticks(fontsize=12)  
plt.title("Top 10 Important Features", fontsize=18)
plt.gca().invert_yaxis()  
plt.tight_layout()  
plt.show()
```
